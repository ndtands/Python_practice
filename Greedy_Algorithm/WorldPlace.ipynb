```
def WorldPlace(k,A):
    G=0
    while(True):
        A.sort(reverse=True)
        Min = A[k-1]
        Allow = int((Min+9999)/10000)
        G =G+Allow
        if (Min <= 0):
            break
        for i in range(0,k):
            A[i]=A[i]-Allow
    return G
```

       A = { 4,4,4,4,4 }

```
A= [ 4,4,4,4,4 ]
print (A)
k=4
print(k)
print(WorldPlace(k,A))
```
